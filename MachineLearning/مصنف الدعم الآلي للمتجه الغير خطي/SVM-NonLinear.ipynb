```python
import matplotlib.pyplot as plt
 
from sklearn import datasets
 
circle_X, circle_y = datasets.make_circles(n_samples=300, noise=0.05)
```

```python
print("أول 5 صفوف من circle_X (الإحداثيات):")
print(circle_X[:10])
print("\nأول 5 قيم من circle_y (التصنيفات):")
print(circle_y[:10])
```

```python
print(circle_X)
```

```python
import matplotlib.pyplot as plt

# رسم البيانات
plt.scatter(circle_X[circle_y == 0, 0], circle_X[circle_y == 0, 1], c='red', label='0')
plt.scatter(circle_X[circle_y == 1, 0], circle_X[circle_y == 1, 1], c='blue', label='1')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Circles Dataset')
plt.legend()
plt.show()
```

```python
plt.scatter(circle_X[:, 0], circle_X[:, 1], c=circle_y, marker='.')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(circle_X, circle_y)

from sklearn import svm
model = svm.SVC(kernel='rbf')

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score

print('\nAccuracy: {:.2f}\n'.format(100*accuracy_score(y_test, y_pred)))

print('\nPrecision: {:.2f}\n'.format(100*precision_score(y_test, y_pred)))

print('\nRecall: {:.2f}\n'.format(100*recall_score(y_test, y_pred)))

print('\nF1: {:.2f}\n'.format(100*f1_score(y_test, y_pred)))
```
