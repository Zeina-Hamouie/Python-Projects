```python
import matplotlib.pyplot as plt

# رسم المخطط الشريطي
ax = df.gender.value_counts().plot(kind='bar')

# إضافة القيم على الأعمدة يدويًا
for p in ax.patches:  # التكرار عبر الأعمدة
    height = p.get_height()  # الحصول على ارتفاع العمود
    ax.text(p.get_x() + p.get_width() / 2, height, 
            f'{height}', ha="center", va="bottom")  # إضافة الرقم على قمة العمود

# إضافة العناوين
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# عرض المخطط
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.countplot(data=df, x="gender", hue="noshow")  # ملاحظة: لا داعي لاستخدام df.gender
for label in ax.containers:
    ax.bar_label(label)

ax.set_title("Show/NoShow for Females and Males")

ax.set_xticks([0, 1])  # ضبط مواضع العلامات
x_ticks_labels = ['Female', 'Male']
ax.set_xticklabels(x_ticks_labels)  # ضبط التسميات
plt.xlabel('Gender')
plt.ylabel('Number')
plt.show()
```
