```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the count of outcomes (diabetes vs no diabetes)
plt.figure(figsize=(8, 5))
sns.countplot(x='Outcome', data=df)
plt.title('Count of Diabetes Outcomes')
plt.xlabel('Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Check data types and missing values
print("\nData Types and Missing Values:")
display(df.info())

# Descriptive statistics for numerical features
print("\nDescriptive Statistics:")
display(df.describe())

# Visualize distributions of numerical features
print("\nHistograms of Numerical Features:")
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.columns[:-1]):  # Exclude 'Outcome'
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()

print("\nBox Plots of Numerical Features:")
plt.figure(figsize=(25, 15))
for i, col in enumerate(df.columns[:-1]):  # Exclude 'Outcome'
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()

# Correlation between numerical features and target variable
print("\nCorrelation Matrix:")
correlation_matrix = df.corr()
display(correlation_matrix)

print("\nHeatmap of Correlation Matrix:")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
